```python
import requests
from bs4 import BeautifulSoup
url_1="https://www.naukri.com/ml-engineer-jobs-in-india?k=ml+engineer&l=india"
url_2="https://www.naukri.com/ml-engineer-jobs-in-india-2"
url_3="https://www.naukri.com/ml-engineer-jobs-in-india-3"
url_4="https://www.naukri.com/ml-engineer-jobs-in-india-4"
url_5="https://www.naukri.com/ml-engineer-jobs-in-india-5"
res1=requests.get(url_1)
print(res1)
```

However the response object returned by HTTP request of requests library doesnâ€™t have HTML tags, 

```python
soup1=BeautifulSoup(res1.text,'html.parser')
print(soup1)
```

So to get the HTML source of web page we shall use selenium library of Python which is used for automation. I am using Chrome webdriver here.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
import pprint

# Specify the full path to the ChromeDriver executable
chrome_service = Service('chromedriver.exe')

# Create Chrome instance with options and service
chrome_browser = webdriver.Chrome(service=chrome_service)
chrome_browser.maximize_window()
#Opens chrome browser and redirects to the link provided in the get method.
chrome_browser.get(url_1)
time.sleep(5)

soup1 = BeautifulSoup(chrome_browser.page_source,'html.parser')
print(soup1.prettify())
```

```python
chrome_browser.get(url_2)
time.sleep(5)

soup2 = BeautifulSoup(chrome_browser.page_source,'html.parser')
print(soup2.prettify())
```

```python
job_list_container= job_list_container1 + job_list_container2 + job_list_container3 + job_list_container4 + job_list_container5
```

```python
# Initialize lists to store job details
job_titles = []
job_locations = []
company = []
key_skills = []

# Initialize an empty list to store individual skills
individual_skills = []

for job in job_list_container:
    try:
        # Extract job title
        title = job.find("a", class_="title").text.strip()
        if title.lower() in ['ml engineer','ml engineer - junior','ml engineer/data scientist', 'ml engineer i', 'ml / data scientist','ml engineer/data scientist','al/ml engineer']:
            job_titles.append('ML Engineer')
        elif title.lower() in ['sr ml engineer','sr. ml engineer', 'senior ml engineer', 'senior ml engineer - Mlops','sr . ml engineer', 'lead ml engineer','ml engineer ii']:
            job_titles.append('Sr. ML Engineer')
        elif title.lower() in ['data scientist','data analyst','data analyst / data scientist','data analyst/data scientist', 'data analyst / scientist 3', 'data scientist -chennai', 'data scientist - ai/ml','software developer - data science','business analyst /data analyst /data scientist','data scientist - ml, python,  - 4-7 years -mumbai/pune/bangalore']:
            job_titles.append('Data Analyst/Data Scientist')  
        elif title.lower() in ['ml Engineer & ml ops', 'ml ops engineer']:
            job_titles.append('ML Ops Engineer')   
        elif title.lower() in ['senior ml ops engineer', 'senior ml ops engineer - wiq']:
            job_titles.append('Sr. ML Ops Engineer')       
        else:
            job_titles.append(title)
    except:
        job_titles.append("None")
    print(title)
    try:
        # Extract job location
        location = job.find("span", class_="locWdth").text.strip()
        if location.lower() in ['bangalore/bengaluru','bengaluru']:
            job_locations.append('Bangalore/Bengaluru')
        elif location.lower() in ['chennai','chennai, tamil nadu']:
            job_locations.append('Chennai')   
        elif location.lower() in ['gurugram, haryana','gurgaon/gurugram']:
            job_locations.append('Gurugram') 
        elif location.lower() in ['hybrid - bangalore/bengaluru','hybrid - bengaluru']:
            job_locations.append('Hybrid - Bangalore/Bengaluru') 
        elif location.lower() in ['mumbai','thane','thane, maharashtra']:
            job_locations.append('Mumbai') 
        elif location.lower() in ['hyderabad, telangana','hyderabad']:
            job_locations.append('Hyderabad')    
        else:
            job_locations.append(location)

    except:
        job_locations.append("None")
    print(location)  
    try:
        # Extract name of the company
        comp_name = job.find("a", class_="comp-name").text.strip()
        company.append(comp_name)
    except:
        company.append("None")
    print(comp_name)
    try:
        # Extract key skills
        skills = job.find('ul', class_='tags-gt').find_all('li', class_='dot-gt tag-li')
        tags_list = [li.text.strip() for li in skills]
        individual_skills.extend(tags_list)
        key_skills.append(",".join(tags_list))
    except:
        key_skills.append("None")
    print(tags_list)

print(len(job_titles))   
print(len(job_locations))   
print(len(company))  
print(len(key_skills))   

```

```python
import pandas as pd
# Create a DataFrame to store the extracted data
job_data = pd.DataFrame({
    "Job Title": job_titles,
    "Location": job_locations,
    "Company": company,
    "Key Skills": key_skills
})

job_data.shape
```
